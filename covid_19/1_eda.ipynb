```python
df = pd.read_csv('data\\raw\\covid_data_India _20th_jun_2021.csv')
```

```python
matplotlib.rcParams["figure.figsize"] = (20,10)

import math

barWidth = 0.3
# fig = plt.subplots(figsize =(20, 10))
fig = plt.figure()
ax = fig.add_subplot(111)

states = df['State/UT']
no_of_confirmed_cases = df['No.of Confirmed cases']
no_of_cured = df['Cure/Discharged/Migrated']
no_of_death = df['No. of Deaths']

# Set position of bar on X axis
br1 = np.arange(len(no_of_confirmed_cases))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

ax.bar(br1, no_of_confirmed_cases, color ='yellow', edgecolor ='grey', width = barWidth, label = 'No. Of Confirmed Case')
ax.bar(br2, no_of_cured, color ='green', edgecolor ='grey', width = barWidth,label = 'No. Of Cured')
ax.bar(br3, no_of_death, color ='red', edgecolor ='grey', width = barWidth, label = 'No. Of Death')

# Adding Xticks
plt.xlabel('States', fontweight ='bold', fontsize = 15)
plt.ylabel('Case Details', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(states))],states, rotation=45, ha='right')

# low = min(no_of_confirmed_cases)
# high = max(no_of_confirmed_cases)
# print(low, high)

# plt.ylim([10000, 8000000])
# plt.ylim(top=50000) #ymax is your value
# plt.ylim(bottom=5000) #ymin is your value

plt.legend()
# plt.show()
# plt.tight_layout()
```
